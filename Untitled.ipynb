```python
import numpy as np
import pandas as pd
%matplotlib inline

from statsmodels.tsa.seasonal import seasonal_decompose
import statsmodels.api as sm
```

```python
df_ship = pd.read_csv('case_shipments.csv')
df_index = pd.read_csv('index.csv')
```

#### Clean df_ship and df_index

```python
# df_ship.drop(['Unnamed: 4'], axis=1, inplace= True)
# df_ship['month'] = pd.to_datetime(df_ship['month'], errors='coerce')
# df_ship.set_index('month', inplace=True)
df_ship['case_wo_pack'] = df_ship['case_shipments'] - 0.2 * df_ship['con_packs']
```

```python
df_index.set_index('month', inplace=True)
```

```python
#creating the train and validation set
train = df_ship[df_ship.index > '1983-12-01'][:int(0.8*(len(df_ship[df_ship.index > '1983-12-01'].index)))]
valid = df_ship[df_ship.index > '1983-12-01'][int(0.8*(len(df_ship[df_ship.index > '1983-12-01'].index))):]

#fit the model
from statsmodels.tsa.vector_ar.var_model import VAR

model = VAR(endog=train)
model_fit = model.fit()

# make prediction on validation
prediction = model_fit.forecast(model_fit.y, steps=len(valid))
```
